```python
import pandas as pd
import os
```

```python
vine_table_data = os.path.join('../Amazon_Vine_Analysis/vine_table.csv')
```

```python
vine_table_df = pd.read_csv(vine_table_data)
```

```python
vine_table_df.head()
```

```python
#1 total_votes count is equal to or greater than 20
new_df = vine_table_df.loc[(vine_table_df["total_votes"] >= 20)]
new_df
```

```python
#2 where the number of helpful_votes divided by total_votes is equal to or greater than 50%
new_df2 = new_df.loc[(new_df["helpful_votes"]/new_df["total_votes"] >= 0.5)]
new_df2
```

```python
#3 retrieves all the rows where a review was written as part of the Vine program (paid), vine == 'Y'
new_df3 = new_df2.loc[(new_df2["vine"] == 'Y')]
new_df3
```

```python
#4 retrieve all the rows where the review was not part of the Vine program (unpaid), vine == 'N'
new_df4 = new_df3.loc[(new_df3["vine"] == 'N')]
new_df4
```
